```python
import ollama 
import pandas as pd
from ollama import Client
```

```python
train_df = pd.read_csv('Training data.csv')
```

```python
train_df
```

```python
train_df.columns
```

```python
modelfile = '''
FROM llama2
SYSTEM You are a calendar assistant. Help me with scheduling. 
'''
model = ollama.create(model='test', modelfile=modelfile)
```

```python
client = Client(host='http://localhost:11434')
```

```python
ollama_dataset = [
    {'role': 'user', 'content': row['Query '], 'intent': row['Intent '], 'entity': row['Entities']}
    for _, row in train_df.iterrows()
]
ollama_dataset
```

```python
#ollama.pull("llama2")
```

```python
ollama.chat(model='llama2', messages=ollama_dataset)
```

```python
query = "I want to schedule a meeting with John tomorrow at 10am"
response = ollama.generate(model='llama2', prompt=query)
print(f"response: {response}")
```
